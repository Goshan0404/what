```python
from typing import List
import numpy as np


def average_precision(closest: List[str], target: List[str]):
    # print(f'closest: {closest}', f'target: {target}')
    ids = np.arange(len(closest))
    is_relevant = np.array([1 if pred in target else 0 for pred in closest])


    cumulative_sum = np.cumsum(is_relevant)
    precision_at_k = cumulative_sum / (ids + 1)
    return np.sum(precision_at_k * is_relevant) / len(target)
```

```python
import torch
from transformers import AutoModelForSequenceClassification, AutoTokenizer

tokenizer_reranker = AutoTokenizer.from_pretrained('BAAI/bge-reranker-v2-m3')
model_reranker = AutoModelForSequenceClassification.from_pretrained('BAAI/bge-reranker-v2-m3', torch_dtype=torch.float16).to(device)
model_reranker = model_reranker.half()
model_reranker.eval()

pairs = [['what is panda?', 'hi'], ['what is panda?', 'The giant panda (Ailuropoda melanoleuca), sometimes called a panda bear or simply panda, is a bear species endemic to China.']]
with torch.no_grad():
    inputs = tokenizer_reranker(pairs, padding=True, truncation=True, return_tensors='pt', max_length=512).to(device)
    scores = model_reranker(**inputs, return_dict=True).logits.view(-1, ).float()
    print(scores)
```

```python
def index_db(collection_name_tfidf, description_tfidf, collection_name_semantic, description_semantic, best_args, tokenizer, model, tfidf_vectorizer, chroma_client):
    split_func = lambda x: globals()[best_args["split_function"]](x, **best_args["split_args"])

    articles_splitted = split_articles(articles, split_func)

    tfidf_metadata = []
    for split, id in tqdm(articles_splitted):
        tfidf_metadata.append({"id": str(id), "text": split})

    print("===== Creating embeddings =====")

    texts = [text for text, _ in articles_splitted]
    tf_embeddings = tfidf_vectorizer.fit_transform(texts).toarray()

    sem_embeddings, metadata = extract_embeddings(
    articles_splitted,
    globals()[best_args["extract_function"]],
    tokenizer,
    model
)

    print("===== Creating collections =====")
    chroma_client.create_collection(collection_name_tfidf, description_tfidf)
    chroma_client.create_collection(collection_name_semantic, description_semantic)

    ids = [f"doc_{i}" for i in range(tf_embeddings.shape[0])]
    chroma_client.add_documents(collection_name_tfidf, ids, tf_embeddings, tfidf_metadata)
    chroma_client.add_documents(collection_name_semantic, ids, sem_embeddings, metadata)
```

```python
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer
from chromadb_collection import *
from clean_documents_body import *
from split_data import *
from data import *
from tqdm import tqdm
import torch


def reciprocal_rank_fusion(ranked_lists, k=60):
    scores = {}
    for ranked_list in ranked_lists:
        for rank, doc_id in enumerate(ranked_list):
            if doc_id not in scores:
                scores[doc_id] = 0
            scores[doc_id] += 1 / (k + rank + 1)
    return sorted(scores.keys(), key=lambda x: scores[x], reverse=True)


collection_name_tfidf = "articles_collection_tfidf"
description_tfidf = "Collection contains tf-idf embeddings of articles"

collection_name_semantic = "articles_collection_semantic"
description_semantic = "Collection contains semantic embeddings of documents"

best_args = evaluation_df.iloc[0]

tokenizer, model = load_model(best_args["model"])

tfidf_vectorizer = TfidfVectorizer(preprocessor=TFIDFPreprocessor().preprocess)

# split_func = lambda x: globals()[best_args["split_function"]](x, **best_args["split_args"])
# articles_splitted = split_articles(articles, split_func)
# texts = [text for text, _ in articles_splitted]
# tfidf_vectorizer.fit(texts)

chroma_client = ChromaDBClient()

index_db(collection_name_tfidf, description_tfidf, collection_name_semantic, description_semantic, best_args, tokenizer, model, tfidf_vectorizer, chroma_client)

ap = []
k = 10
n_retrieve = 30

print("===== Evaluating MAP on calibration =====")
for index, query in tqdm(calibration.iterrows(), total=len(calibration)):
    query_tfidf = tfidf_vectorizer.transform([query["body_cleanned"]]).toarray()[0]
    query_embedding_semantic = globals()[best_args["extract_function"]](
        query["body_cleanned"],
        tokenizer=tokenizer,
        model=model
    )

    tfidf_results = chroma_client.get_closest_docs(
        collection_name_tfidf,
        [query_tfidf],
        n_closest=n_retrieve
    )
    semantic_results = chroma_client.get_closest_docs(
        collection_name_semantic,
        [query_embedding_semantic],
        n_closest=n_retrieve
    )

    tfidf_metas = tfidf_results['metadatas'][0]
    semantic_metas = semantic_results['metadatas'][0]

    tfidf_ids = list(dict.fromkeys(item['id'] for item in tfidf_metas))
    semantic_ids = list(dict.fromkeys(item['id'] for item in semantic_metas))

    chunk_text_map = {}
    for item in tfidf_metas + semantic_metas:
        if item['id'] not in chunk_text_map:
            chunk_text_map[item['id']] = item['text']

    combined_closest = reciprocal_rank_fusion(
        [tfidf_ids, semantic_ids], k=60
    )[:k * 2]

    reranking_pairs = [
        [query["body_cleanned"], chunk_text_map[doc_id]]
        for doc_id in combined_closest
    ]

    with torch.inference_mode():
        inputs = tokenizer_reranker(
            reranking_pairs, padding=True, truncation=True,
            return_tensors='pt', max_length=512
        ).to(device)
        scores = model_reranker(**inputs, return_dict=True).logits.view(-1,).cpu().float().numpy()

    top_idx = np.argsort(scores)[::-1]
    id_close = [combined_closest[int(i)] for i in top_idx]

    ap.append(average_precision(id_close, query["ground_truth"].split()))

map = np.mean(ap)
print(f"MAP: {map}")
```
